```python
import numpy as np
import matplotlib.pyplot as plt
import math

a0 = 0
lna0 = 0
a1 = 0
a2 = 0
sis1 = []
sis2 = []
sis3 = []
u0 = 0
u1 = 0
u2 = 0
u3 = 0
u4 = 0
u5 = 0
u6 = 0
u7 = 0
u8 = 0
g0 = []
g1 = []
g2 = []
fx = []
diametro = [1,2,3,1,2,3,1,2]
inclinacao = [0.001, 0.001, 0.001, 0.01, 0.01, 0.01, 0.05, 0.05]
vazao = [1.4, 8.3, 24.2, 4.7, 28.9, 84.0, 11.1, 200.0]

def g0fill(): #Função para preencher G0
    for i in range (8):
        J = 1
        g0.append(J)
    return g0

def g1fill():  #Função para preencher G1
    for i in range (8):
        J = np.log(diametro[i])
        g1.append(J)
    return g1

def g2fill():  #Função para preencher G2
    for i in range (8):
        J = np.log(inclinacao[i])
        g2.append(J)
    return g2

def fxfill():  #Função para preencher fx
    for i in range (8):
        J = np.log(vazao[i])
        fx.append(J)
    return fx

fxfill()
g0fill()
g1fill()
g2fill()

for i in range (8):
    u0 += g0[i]*g0[i]
    
for j in range (8):
    u1 += g1[j]*g0[j]

for k in range (8):
    u2 += g2[k]*g0[k]
    
for l in range (8):
    u3 += g1[l]*g1[l]
    
for m in range (8):
    u4 += g2[m]*g1[m]
    
for n in range (8):
    u5 += g2[n]*g2[n]
    
for o in range (8):
    u6 += fx[o]*g0[o]
    
for p in range (8):
    u7 += fx[p]*g1[p]
    
for q in range (8):
    u8 += fx[q]*g2[q]
    
#print ("u0 = ", u0)
#print ("u1 = ", u1)
#print ("u2 = ", u2)
#print ("u3 = ", u3)
#print ("u4 = ", u4)
#print ("u5 = ", u5)
#print ("u6 = ", u6)
#print ("u7 = ", u7)
#print ("u8 = ", u8)

sis1 = [u0, u1, u2] # Sistema matricial (AL)
sis2 = [u1, u3, u4]
sis3 = [u2, u4, u5]

#print ('Sis1 = ', sis1)
#print ('Sis2 = ', sis2)
#print ("Sis3 = ", sis3)

def l2l(): #Função p/ zerar sis2[0]
    sis1a = sis1
    sis2a = sis2
    for i in range (3):
        sis2a[i] = sis2a[i]*(8/4.276666119016055)
    for j in range (3):
        sis2a[j] = sis2a[j]-sis1a[j]
    return sis2a

sis2 = l2l()

def l3l():  #Função p/ zerar sis3[0]
    sis1a = sis1
    sis3a = sis3
    for i in range (3):
        sis3a[i] = (sis3a[i]*(8/-40.530240942018665))
    for j in range (3):
        sis3a[j] = sis3a[j]-sis1a[j]
    return sis3a

sis3 = l3l()
#print ("Sis3 intermediário", sis3)

def l3ll(): #Função p/ zerar sis3[1]
    sis2a = sis2
    sis3a = sis3
    for i in range (3):
        sis3a[i] = (sis3a[i]*(2.935039178692354/0.20490143583455733))
    for j in range (3):
        sis3a[j] = sis3a[j]-sis2a[j]
    return sis3a

sis3 = l3ll()

#print ()
#print ('Sis1 = ', sis1)
#print ('Sis2 = ', sis2)
#print ("Sis3 = ", sis3)
        
def res():
    a2 = u8/sis3[2]
    a1 = (u7-(sis2[2]*a2))/(sis2[1])
    lna0 = (u6-(sis1[1]*a1)-(sis1[2]*a2))/(sis1[0])
    return lna0, a1, a2

lna0, a1, a2 = res()
#print ("lna0 = ", lna0)

e = math.exp(1)

a0 = e**(lna0)

#print (a0, a1, a2)

print ("Trabalho Referente à segunda nota da disciplina de Cálculo Numérico")
print ("Alunos: Caio Bertoldo, Carlos Henrique, Gabriel Lopes, Kevin Sanches")
print ("Exercício 7.12")
print ()
print ()
print ()
print ("Em um estudo, foi determinada a vazão de água em uma tubulação.")
print ("Descobriu-se que a mesma está relacionada com a inclinação e o diâmetro da tubulação.")
print ("Os dados empíricos do estudo são:")
print ()
print ("Diâmetro    Inclinação    Vazão")
print ("    1         0.001        1.4")
print ("    2         0.001        8.3")
print ("    3         0.001        24.2")
print ("    1         0.01         4.7")
print ("    2         0.01         28.9")
print ("    3         0.01         84.0")
print ("    1         0.05         11.1")
print ("    2         0.05         200.0")
print ()
print ("O estudo sugere que uma equação, regendo a vazão da água, teria a seguinte forma: Q = a0*(D^a1)*(S^a2)")
print ()
print ("Para utilizar o método dos mínimos quadrados, é preciso fazer a linearização da equação")
print ("A equação, linearizada, toma a forma de: lnQ = lna0 + a1*lnD + a2*lnS")
print ("Tomando:")
print ("F(x) = lnQ")
print ("a0 = lna0")
print ("a1 = a1")
print ("a2 = a2")
print ("g0 = 1")
print ("g1 = lnD")
print ("g2 = lnS")
print ()
print ("Desta maneira, podemos construir a matriz, da seguinte maneira:")
print ()
print ('''
|(g0,g0)    (g1,g0)    (g2,g0)|  |lna0|     |(F(x),g0)|
|(g0,g1)    (g1,g1)    (g2,g1)|  | a1 |  =  |(F(x),g1)|
|(g0,g2)    (g1,g2)    (g2,g2)|  | a2 |     |(F(x),g2)|
''')
print ()
print ("Resolvendo a matriz, chegamos nos valores de:")
print()
print ("lna0 = ", lna0)
print ("a1 = ", a1)
print ("a2 = ", a2)
print ()
print ("Para que possamos obter o valor de a0, aplicamos o 'e' em lna0. Desta maneira, obtemos o valor de a0 = ", a0)
print ()
print ("Portanto, a equação que rege a vazão da água é: ", a0, "*D^(", a1, ")*S^(", a2, ")")
print ()
print ("A vazão de água em uma tubulação com diâmetro de 2.5m e uma inclinação de 0.0025 é:")
print ("Q = ", a0*(2.5**a1)*(0.0025**a2), "m³/s")
```
